```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20,10))
```

Por Barra

```python
fig = grupo_PROCESADOR['CANON MENSUAL'].mean().plot.bar(color = 'green')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor promedio del Alquiler', {'fontsize': 22})
```

Por Linea

```python
fig = grupo_PROCESADOR['CANON MENSUAL'].max().plot.line(color = 'blue')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor maximo por AREA', {'fontsize': 22})
```
